```python
############################################
################### AT HOME ######################
############################################
def repeat(L, n):
    """ L is a list of ints
        n is a positive int
    Mutates L to contain whatever elements L has right now repeated n times. """
    # your code here 
    print(n)
    L_copy = L[:]
    for i in range(n-1):
        print(i)
        # L += L  as this is mutating old list
        L = L + L_copy
    return L

    
Lin = [1,2,3]
repeat(Lin, 3)
print(Lin)    # prints [1, 2, 3, 1, 2, 3, 1, 2, 3]
```

In the code you provided, the line L = L + L_copy does not mutate the original list L. Instead, it creates a new list and assigns it to L, which only changes the local variable L inside the function. This is why the original list Lin remains unchanged outside the function.

```python
############################################
################### AT HOME ######################
############################################
def repeat(L, n):
    """ L is a list of ints
        n is a positive int
    Mutates L to contain whatever elements L has right now repeated n times. """
    # your code here 
    print(n)
    L_copy = L[:]
    for i in range(n-1):
        print(i)
        # L += L  as this is mutating old list
        # L = L + L_copy
        L += L_copy
    return L

    
Lin = [1,2,3]
repeat(Lin, 3)
print(Lin)    # prints [1, 2, 3, 1, 2, 3, 1, 2, 3]
```
